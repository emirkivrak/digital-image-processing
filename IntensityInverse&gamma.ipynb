```python
import numpy as np
from skimage import data, io, filters
import matplotlib.pyplot as plt
```

```python
def convert_rgb_to_gray_level(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]
    im_2=np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            im_2[i,j]=get_distance(im_1[i,j,:])
    return im_2

def get_distance(v,w=[1/3,1/3,1/3]):
    a,b,c=v[0],v[1],v[2]
    w1,w2,w3=w[0],w[1],w[2]
    d=((a**2)*w1+
    (b**2)*w2+
    (c**2)*w3)**.5
    return d
```

```python

def my_f_1(a,b):
    assert a>=0;" intensity pozitive", "error intensity not pozitive"
    if(a<=255-b):
        return a+b
    else:
        return 255
my_f_1(243,30)

def my_f_2(a):
    #assert a>=0;" intensity pozitive", "error intensity not pozitive"
    return int(255-a)
my_f_2(243)
```

```python
img_1=convert_rgb_to_gray_level(img_0)
plt.imshow(img_1,cmap="gray")
plt.show()
```

```python
m,n = img_1.shape[0],img_1.shape[1]
img_2 = np.zeros((m,n),dtype="uint8")

for i in range(m):
    for j in range(n):
        intensity = img_1[i,j]
        img_2[i,j] = my_f_2(intensity)
        
```
